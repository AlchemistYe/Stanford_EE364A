```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solver(obj, constraints):
    prob = cp.Problem(obj, constraints)
    # The optimal objective value is returned by `prob.solve()`.
    prob.solve()
    #print("status:", prob.status, "\noptimal value", prob.value)
    return prob.value
```

```python
np.random.seed(0)
n = 100
m = 300
A = np.random.rand(m, n)
b = A @ np.ones((n,))/2
c = -np.random.rand(n, 1)
```

```python
x = cp.Variable(n)
constraints = [A@x <= b,
               x >= 0,
               x <= 1]

objective = cp.Minimize(c.T@x)
L = solver(objective, constraints)
```

```python
thresh = np.linspace(0, 1, 101)
obj, maxviol = [], []
for t in thresh:
    x_round = x.value >= t
    obj += [(c.T @ x_round)[0]]
    maxviol += [np.max(A @ x_round - b)]

maxviol = np.array(maxviol)
obj = np.array(obj)
feasi_t = maxviol <= 0
infeasi_t = maxviol > 0
```

```python
plt.subplot(2, 1, 1)
plt.plot(thresh[feasi_t], maxviol[feasi_t], c = 'b', label = "feasible")
plt.plot(thresh[infeasi_t], maxviol[infeasi_t], c = 'r', label = "infeasible")
plt.legend()
plt.xlabel("t")
plt.ylabel("max violation")

plt.subplot(2, 1, 2)
plt.plot(thresh[feasi_t], obj[feasi_t], c = 'b', label = "feasible")
plt.plot(thresh[infeasi_t], obj[infeasi_t], c = 'r', label = "infeasible")
plt.legend()
plt.xlabel("t")
plt.ylabel("objective")
```

```python
U = np.min(obj[feasi_t])
print("L:", L, "U:", U, "\nU - L:", U-L)
```

```python
thresh_return = np.linspace(0, 0.3, 50)
constraints = [cp.sum(x) == 1]
obj = cp.Minimize(x.T@S@x)
risk1 = []
risk2 = []

for thresh in thresh_return:
    constraints1 = constraints + [pbar.T@x == thresh, x >= 0]
    constraints2 = constraints + [pbar.T@x == thresh, cp.sum(cp.neg(x)) <= 0.5]
    prob1 = solver(obj, constraints1)
    prob2 = solver(obj, constraints2)
    risk1.append(prob1)
    risk2.append(prob2)

plt.plot(risk1, thresh_return, c = 'b', label = "long-only")
plt.plot(risk2, thresh_return, c = 'r', label = "limit on total short position")
plt.xlabel("standard deviation")
plt.ylabel("mean return")
plt.legend()
plt.show()
```
