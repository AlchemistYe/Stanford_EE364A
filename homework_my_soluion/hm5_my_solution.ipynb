## Minimax rational fit to the exponential

```python
l_b = 0
u_b = np.exp(3)
k = 201
acc = 1e-3

a = cp.Variable(3)
b = cp.hstack((1, cp.Variable(2)))
ti = np.linspace(-3,3,k)
yi = np.exp(ti).reshape(1, -1)
matrix_y = np.concatenate((yi, yi, yi))
gamma = cp.Parameter(nonneg=True)
matrix_t = np.concatenate((np.ones(k).reshape(1,-1), ti.reshape(1,-1), (ti**2).reshape(1,-1)))
```

```python
def prob_func(gamma):
    constraints = [cp.abs(a@matrix_t - b@(matrix_y*matrix_t)) <= gamma*b@matrix_t]
    obj = cp.Minimize(gamma)
    prob = cp.Problem(obj, constraints)
    return prob

def check_whether_optimal_in(upper, lower):
    gamma1 = cp.Parameter(nonneg=True)
    gamma1.value = upper
    prob = prob_func(gamma1)
    prob.solve()
    if not prob.status in ['optimal', 'optimal_inaccurate']:
        raise Exception("upper bound is too small and no gamma exists")

    gamma1.value = lower
    prob = prob_func(gamma1)
    prob.solve()
    if prob.status in ['optimal', 'optimal_inaccurate']:
        raise Exception("lower bound is too big and gamma is not optimal")
    
check_whether_optimal_in(u_b, l_b)
max_iter = 1000
iter_num = 0
while u_b - l_b >= acc and iter_num <= max_iter:
    iter_num += 1
    gamma.value = (u_b + l_b)/2
    prob = prob_func(gamma)
    prob.solve()
    status = prob.status
    if status in ['optimal', 'optimal_inaccurate']:
        u_b = gamma.value
        optimal_a = a.value
        optimal_b = b.value
        optimal_gamma = gamma.value
    else:
        l_b = gamma.value

if u_b - l_b >= acc and iter_num >= max_iter:
    print("solution can't converge")

print("optimal value of a is: {} \noptimal value of b is: {} \noptimal objective value is: {:.3f}:"
      .format(optimal_a, optimal_b[1:], optimal_gamma))
```

```python
optimal_rational = optimal_a@matrix_t/(optimal_b@matrix_t)
error = optimal_rational - yi.reshape(201)
plt.plot(ti, optimal_rational, color = 'r')
plt.xlabel("t_i")
plt.ylabel("optimal rational fit")
```

## Worst-case probability of loss

```python
r = np.linspace(-30, 70, 100)
n = 100
mu_1 = 8 
mu_2 = 20
sigma_1 = 6
sigma_2 = 17.5
rho = -0.25
p1 = np.exp(-(r - mu_1)**2/(2 * sigma_1**2))
p1 = p1/np.sum(p1)
p2 = np.exp(-(r - mu_2)**2/(2 * sigma_2**2))
p2 = p2/np.sum(p2)
neg_mask = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if r[i] + r[j] <= 0:
            neg_mask[i][j] = 1

P = cp.Variable((n, n), nonneg=True)
constraints = [P@np.ones(n) == p1,
               P.T@np.ones(n) == p2,
               (r - mu_1*np.ones(n)).T@P@(r - mu_2*np.ones(n)) == rho*sigma_1*sigma_2]
obj = cp.Maximize(cp.sum(cp.multiply(P, neg_mask)))
prob = cp.Problem(obj, constraints)
p_loss = prob.solve()
p_loss
```
